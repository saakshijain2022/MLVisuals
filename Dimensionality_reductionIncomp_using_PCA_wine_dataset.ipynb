```python
import numpy as np 
import pandas as pd
import matplotlib.pyplot as plt

data_set = pd.read_csv('./DATA_SETS/wine_dimensional_reduction.data', sep=',',names=["Customer","Alcohol","Malic_acid","Ash","Ash_Alcan","Magnesium","Total_Phenols","Flavanoids","Nonflav","Proanthocyan","Color","Hue","OD280","Proline"])
print(type(data_set))
print(data_set)
```

```python
# principal component analysis is the technique used to map our features in some different domain/space
# this PCA technique is possible to reduce the dimensions 

# customer (target variable) has 1,2,3 in it so can go by classification 

# step 3 : extracting the predictorsand target variables
# predictors = ['Gender','Age', 'EstimatedSalary']
# target = ['Purchased']

# x = data_set[predictors].values
# y = data_set[target].values


# this approach is not appropiate for this cuz we had so many columns so time consumed to write their names
```

```python
# step 3 : extracting the predictorsand target variables
x= data_set.iloc[:,1:14]
# [:,1:14]   mena s all rows ,columns  1 to 14(not included)
# 14 will be ignored

y=data_set.iloc[:,0]
# [:,0]    all rows and first(0) column
print(x)
print("target variable")
print(y)
```

```python
# step 4: splitting the dataset
from sklearn.model_selection import train_test_split
x_train ,x_test ,y_train ,y_test = train_test_split(x,y,test_size=0.2,random_state =0)
```

```python
# step 5: Feature Scaling (additional new)
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
# keep x_test/train in the parenthese   and done to get every parameter in a com mon range
x_train = sc.fit_transform(x_train)
x_test = sc.fit_transform(x_test)

print(x_test)
```

```python
# still we have 13 number of colums so though PCA we can reduce these dimensions

# step :dimensionality reduction using PCA
from sklearn.decomposition import PCA

pca = PCA(n_components=2)
# no of pricipal components is <= no of parameters

#  n_components / principal components  how many angles/ views you want to create and are orthogonal components
# reduce no of dimensions
# if we assume 2 components for pca then there will be 2 principal components  are orthogonal to each other 

x_train = pca.fit_transform(x_train)
x_test = pca.fit_transform(x_test)
# why only x cuz x has 13 columns of different ranges of values which has been scaled down by standard scalar
# why only y cuz one columns and is the the common ranges without scaling down

print(x_test)
# we are getting 2 columns unlike 13 in original
```
